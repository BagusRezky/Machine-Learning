## Deskripsi
Pada praktikum kali ini, Anda diminta untuk membuat model ANN untuk mengklasifikasi potensi seorang customer akan meninggalkan perusahaan Anda atau tidak. Istirlah populer dari fenomena ini disebut sebagai 'churn'. Tingkat churn yang tinggi (chrun rate) akan berdampak tidak baik bagi perusahaan.

## Pra Pengolahan Data

## Langkah 1 - Import Library

```python
%pip install tensorflow
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

## Langkah 2 - Load Data

```python
dataset = pd.read_csv('Dataset/Churn_Modelling.csv')
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

## Langkah 3 - Encoding Data Kategorikal

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
```

```python
print(X)
```

## Langkah 4 - Encoding Kolom "Geography" dengan One Hot Encoder

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

## Langkah 5 - Split Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Langkah 6 - Scaling Fitur

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Langkah 7 - Membuat Model ANN

## Inisisasi ANN

```python
ann = tf.keras.models.Sequential()
```

## Langkah 8 - Membuat Input Layer dan Hidden Layer Pertama

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

## Langkah 9 - Membuat Hidden Layer Kedua

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

## Langkah 10 - Membuat Output Layer

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

## Training Model


## Langkah 11 - Compile (Menyatukan Arsitektur) ANN

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

## Langkah 12 - Fitting Model 

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

## Membuat Prediksi

## Langkah 13 - Modelkan Data baru dan buat prediksi

```python
print(ann.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]])) > 0.5)
```

## Langkah 14 - Prediksi Dengan Data Testing

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Langkah 15 - Cek Akurasi dan Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
